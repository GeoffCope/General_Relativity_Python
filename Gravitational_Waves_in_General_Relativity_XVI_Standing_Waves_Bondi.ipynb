```python
from sympy import init_session
init_session(use_latex=True)
```

```python
import warnings
import matplotlib.cbook
warnings.filterwarnings("ignore",category=matplotlib.cbook.mplDeprecation)
```

```python
t,r,theta,z,dt,dr,dtheta,dz = symbols( 't r theta z dt dr dtheta dz' ) 
```

```python
epsilon,phi,Lambda,mu = symbols('epsilon phi Lambda mu')
```

```python
phi = Function('phi')(r,theta)
phi
```

```python
mu = Function('mu')(r,theta)
mu
```

```python
Lambda = Function('Lambda')(r,theta)
Lambda
```

```python
variables = Matrix( [t,r,theta,z] )
variables
```

```python
differentials = Matrix( [dt,dr,dtheta,dz] ) 
differentials
```

```python
lineElement = expand((1-epsilon**2)*exp(2*phi)*dt**2 + 2*epsilon*r*exp(phi+Lambda)*dt*dtheta - exp(2*mu)*dz**2 - exp(2*Lambda)*(dr**2 + r**2 * dtheta**2))
lineElement
```

```python
lineElement.subs(epsilon,0)
```

```python
g = zeros(4)

for i in range(4):
    for j in range(4):
        if i == j:
            g[i,j] = lineElement.coeff( differentials[i],2 ) 
        else:
            g[i,j] = Rational(1,2)*lineElement.coeff( differentials[i]*differentials[j],1 )
            
g
```

```python
g.subs(epsilon,0)
```

```python
import sympy
```

```python
m = sympy.Array(g)
m
```

```python
m.subs(epsilon,0)
```

```python
from einsteinpy.symbolic import *
```

```python
syms = sympy.symbols("t r theta z")
t,r,theta,z = syms
```

```python
metric = MetricTensor(m,syms)
```

```python
metric.subs(epsilon,0)
```

```python
ch = ChristoffelSymbols.from_metric(metric)
simplify(ch.tensor())
```

```python
ch.tensor().subs(epsilon,0)
```

```python
Ric = RicciTensor.from_metric(metric)
simplify(Ric.tensor())
```

```python
Ric.tensor().subs(epsilon,0)
```

```python
einst = EinsteinTensor.from_metric(metric)
simplify(einst.tensor())
```

```python
einst.tensor().subs(epsilon,0)
```

```python
R = RicciScalar.from_riccitensor(Ric)
R.simplify()
R.expr
```

```python
R.expr.subs(epsilon,0)
```

```python
rm1 = RiemannCurvatureTensor.from_christoffels(ch)
rm1.tensor()
```

```python
rm1.tensor().subs(epsilon,0)
```

```python
weyl = WeylTensor.from_metric(metric)
weyl.tensor()
```

```python
weyl.tensor().subs(epsilon,0)
```
