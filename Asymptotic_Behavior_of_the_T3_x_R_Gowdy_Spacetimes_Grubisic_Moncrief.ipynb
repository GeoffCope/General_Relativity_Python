```python
t,theta,x,y,dt,dtheta,dx,dy = symbols('t theta x y dt dtheta dx dy')
```

```python
W = Function('W')(t,theta)
W
```

```python
Phi = Function('Phi')(t,theta)
Phi
```

```python
A = Function('A')(t,theta)
A
```

```python
variables = Matrix([t,theta,x,y])
variables
```

```python
differentials = Matrix([dt,dtheta,dx,dy])
differentials 
```

```python
lineElement = expand( exp(2*A)*( -exp(2*t)*dt**2 + dtheta**2) + exp(-t)*( cosh(W) + cos(Phi)*sinh(W))*dx**2 + 2*exp(-t)*sin(Phi)*sinh(W)*dx*dy + exp(-t)*( cosh(W)-cos(Phi) * sinh(W))*dy**2  )
lineElement 
```

```python
g = zeros(4)

for i in range(4):
    for j in range(4):
        if i == j:
            g[i,j] = lineElement.coeff( differentials[i],2 ) 
        else:
            g[i,j] = Rational(1,2)*lineElement.coeff( differentials[i]*differentials[j],1 )
            
g
```

```python
import sympy
```

```python
m = sympy.Array(g)
m
```

```python
from einsteinpy.symbolic import *
```

```python
syms = sympy.symbols("t theta x y")
t,theta,x,y = syms
```

```python
metric = MetricTensor(m,syms)
```

```python
ch = ChristoffelSymbols.from_metric(metric)
simplify(ch.tensor())
```

```python
Ric = RicciTensor.from_metric(metric)
Ric.tensor()
```

```python
R = RicciScalar.from_riccitensor(Ric)
R.simplify()
R.expr
```

```python
einst = EinsteinTensor.from_metric(metric)
einst.tensor()
```
