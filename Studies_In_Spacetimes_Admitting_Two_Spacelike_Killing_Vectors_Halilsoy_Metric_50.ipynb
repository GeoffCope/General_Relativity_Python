```python
variables = Matrix([t,rho,phi,z])
variables
```

```python
differentials = Matrix([dt,drho,dphi,dz])
differentials 
```

```python
lineElement = expand( exp(2*(gamma - psi)) * (dt**2 -drho**2) - exp(2*psi) * ( dz + omega * dphi)**2 -rho**2*exp(-2*psi)*dphi**2 ) 
lineElement
```

```python
g = zeros(4)

for i in range(4):
    for j in range(4):
        if i == j:
            g[i,j] = lineElement.coeff( differentials[i],2 ) 
        else:
            g[i,j] = Rational(1,2)*lineElement.coeff( differentials[i]*differentials[j],1 )
            
g
```

```python
import sympy
```

```python
m = sympy.Array(g)
m
```

```python
from einsteinpy.symbolic import *
```

```python
syms = sympy.symbols("t rho phi z")
t,rho,phi,z = syms
```

```python
metric = MetricTensor(m,syms)
```

```python
ch = ChristoffelSymbols.from_metric(metric)
simplify(ch.tensor())
```

```python
Ric = RicciTensor.from_metric(metric)
Ric.tensor()
```

```python
R = RicciScalar.from_riccitensor(Ric)
R.simplify()
R.expr
```

```python
einst = EinsteinTensor.from_metric(metric)
einst.tensor()
```

```python
ricciSimplified = simplify(Ric.tensor())
ricciSimplified
```

```python
einsteinSimplified = simplify(einst.tensor())
einsteinSimplified
```

```python
d2psidt2 = Eq( diff(psi,t,t),expand(solve( ricciSimplified[3,3].args[1], diff(psi,t,t) )[0] )) 
d2psidt2
```

```python
d2omegadt2 = Eq( diff(omega,t,t),solve( ricciSimplified[2,2].args[1].subs(d2psidt2.lhs,d2psidt2.rhs ), diff(omega,t,t))[0] ) 
d2omegadt2 
```

```python
dgammadrho = Eq( diff(gamma,rho) , expand( solve( simplify( ( ricciSimplified[0,0] + ricciSimplified[1,1] ) ), diff(gamma,rho) )[0] ) ) 
dgammadrho 
```

```python
dgammadt = Eq(diff(gamma,t),solve( einsteinSimplified[1,0], diff(gamma,t))[0] ) 
dgammadt
```

```python
#  Introduce auxiliary variables and convert evolution equations to first order 
vacuumFieldEquations = Matrix([d2psidt2,d2omegadt2,dgammadrho,dgammadt])
vacuumFieldEquations
```
