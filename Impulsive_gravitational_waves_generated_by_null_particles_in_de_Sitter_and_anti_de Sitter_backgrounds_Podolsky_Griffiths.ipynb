```python
from sympy import init_session
init_session(use_latex=True)
```

```python
import warnings
import matplotlib.cbook
warnings.filterwarnings("ignore",category=matplotlib.cbook.mplDeprecation)
```

```python
Z_0,Z_1,Z_2,Z_3,Z_4 = symbols( 'Z_0 Z_1 Z_2 Z_3 Z_4' ) 
```

```python
t,r,theta,phi,m,a = symbols('t r theta phi m a')
```

```python
equation2pt2 = Eq( Z_0**2 - Z_1**2 -Z_2**2 -Z_3**2 -Z_4**2 , - a**2 )
equation2pt2
```

```python
equation2pt4a = Eq( Z_0 , (a**2-r**2)**Rational(1,2) * sinh(t/a) ) 
equation2pt4a
```

```python
equation2pt4b = Eq( Z_4 , (a**2-r**2)**Rational(1,2) * cosh(t/a) ) 
equation2pt4b
```

```python
equation2pt4c = Eq(Z_1, r * cos(theta))
equation2pt4c
```

```python
equation2pt4d = Eq(Z_2, r * sin(theta) * cos(phi))
equation2pt4d
```

```python
equation2pt4e = Eq(Z_3, r * sin(theta) * sin(phi))
equation2pt4e
```

```python
equation2pt2.subs(equation2pt4a.lhs,equation2pt4a.rhs)
```

```python
equation2pt2.subs(equation2pt4a.lhs,equation2pt4a.rhs).subs(equation2pt4b.lhs,equation2pt4b.rhs)
```

```python
equation2pt2.subs(equation2pt4a.lhs,equation2pt4a.rhs).subs(equation2pt4b.lhs,equation2pt4b.rhs).subs(equation2pt4c.lhs,equation2pt4c.rhs)
```
