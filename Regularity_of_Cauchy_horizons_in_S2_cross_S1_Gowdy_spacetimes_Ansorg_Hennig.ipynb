```python
t,theta,phi,delta,dt,dtheta,dphi,ddelta = symbols('t theta phi delta dt dtheta dphi ddelta')
```

```python
M = Function('M')(t,theta)
M
```

```python
Q = Function('Q')(t,theta)
Q
```

```python
u = Function('u')(t,theta)
u
```

```python
variables = Matrix([t,theta,phi,delta])
variables
```

```python
differentials = Matrix([dt,dtheta,dphi,ddelta])
differentials 
```

```python
lineElement = expand( exp(M) * (-dt**2 + dtheta**2) + exp(u) * sin(theta)**2 * (dphi + Q * ddelta )**2 + exp(-u) * sin(t)**2 * ddelta**2 )  
lineElement
```

```python
g = zeros(4)

for i in range(4):
    for j in range(4):
        if i == j:
            g[i,j] = lineElement.coeff( differentials[i],2 ) 
        else:
            g[i,j] = Rational(1,2)*lineElement.coeff( differentials[i]*differentials[j],1 )
            
g
```

```python
import sympy
```

```python
m = sympy.Array(g)
m
```

```python
R = RicciScalar.from_riccitensor(Ric)
R.simplify()
R.expr
```

```python
einst = EinsteinTensor.from_metric(metric)
einst.tensor()
```
