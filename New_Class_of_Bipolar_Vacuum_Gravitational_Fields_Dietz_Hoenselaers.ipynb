```python
rho,z,phi,t,drho,dz,dphi,dt = symbols( 'rho z phi t drho dz dphi dt' ) 
```

```python
chi = Function('chi')(rho,z)
chi
```

```python
gamma = Function('gamma')(rho,z)
gamma
```

```python
variables = Matrix( [rho,z,phi,t] )
variables
```

```python
differentials = Matrix( [drho,dz,dphi,dt] ) 
differentials
```

```python
lineElement = expand( -(dt**2*exp(2*chi)) +  exp(-2*chi)*((dz**2 + drho**2)*exp(2*gamma) + dphi**2*rho**2 ) )
lineElement
```

```python
g = zeros(4)

for i in range(4):
    for j in range(4):
        if i == j:
            g[i,j] = lineElement.coeff( differentials[i],2 ) 
        else:
            g[i,j] = Rational(1,2)*lineElement.coeff( differentials[i]*differentials[j],1 )
            
g
```

```python
import sympy
```

```python
m = sympy.Array(g)
m
```

```python
from einsteinpy.symbolic import *
```

```python
syms = sympy.symbols("rho z phi t")
rho,z,phi,t = syms
```

```python
metric = MetricTensor(m,syms)
```

```python
ch = ChristoffelSymbols.from_metric(metric)
ch.tensor()
```

```python
Ric = RicciTensor.from_metric(metric)
Ric.tensor()
```

```python
R = RicciScalar.from_riccitensor(Ric)
R.simplify()
R.expr
```

```python
einst = EinsteinTensor.from_metric(metric)
einst.tensor()
```

```python
ricciSimplifed = simplify(Ric.tensor())
ricciSimplifed
```

```python
einsteinSimplified = simplify(einst.tensor())
einsteinSimplified
```

```python
equation1pt1a = expand((1/rho)*ricciSimplifed[3,3].args[1] ) 
equation1pt1a
```

```python
equation1pt1b1 = Eq(diff(gamma,rho),solve( einsteinSimplified[0,0], diff(gamma,rho) )[0] ) 
equation1pt1b1
```

```python
equation1pt1b2 = Eq(diff(gamma,z) ,solve( einst.tensor()[1,0], diff(gamma,z) )[0] ) 
equation1pt1b2
```

```python
vacuumFieldEquations = Matrix([equation1pt1a,equation1pt1b1,equation1pt1b2])
vacuumFieldEquations
```
