```python
import warnings
import matplotlib.cbook
warnings.filterwarnings("ignore",category=matplotlib.cbook.mplDeprecation)
```

```python
tau,u,eta,zeta,dtau,du,deta,dzeta = symbols( 'tau u eta zeta dtau du deta dzeta' ) 
```

```python
phi = Function('phi')(u)
phi
```

```python
beta = Function('beta')(u)
beta
```

```python
variables = Matrix( [tau,u,eta,zeta] )
variables
```

```python
differentials = Matrix( [dtau,du,deta,dzeta] ) 
differentials
```

```python
lineElement = expand( - exp(2*phi) * du**2 + ( 2 * exp(2*phi) * du * dtau ) - exp(-2*beta) * u**2 * ( dzeta**2 + exp(4*beta) * deta**2 ))
lineElement
```

```python
g = zeros(4)

for i in range(4):
    for j in range(4):
        if i == j:
            g[i,j] = lineElement.coeff( differentials[i],2 ) 
        else:
            g[i,j] = Rational(1,2)*lineElement.coeff( differentials[i]*differentials[j],1 )
            
g
```

```python
import sympy
```

```python
m = sympy.Array(g)
m
```

```python
from einsteinpy.symbolic import *
```

```python
syms = sympy.symbols("tau u eta zeta")
tau,u,eta,zeta = syms
```

```python
metric = MetricTensor(m,syms)
```

```python
ch = ChristoffelSymbols.from_metric(metric)
ch.tensor()
```

```python
Ric = RicciTensor.from_metric(metric)
Ric.tensor()
```

```python
einst = EinsteinTensor.from_metric(metric)
einst.tensor()
```

```python
R = RicciScalar.from_riccitensor(Ric)
R.simplify()
R.expr
```

```python
rm1 = RiemannCurvatureTensor.from_christoffels(ch)
rm1.tensor()
```

```python
weyl = WeylTensor.from_metric(metric)
weyl.tensor()
```
