```python
import warnings
import matplotlib.cbook
warnings.filterwarnings("ignore",category=matplotlib.cbook.mplDeprecation)
```

```python
u,v,x,y,du,dv,dx,dy = symbols( 'u v x y du dv dx dy', positive = True, real = True ) 
```

```python
M = Function('M')(u,v)
M
```

```python
U = Function('U')(u,v)
U
```

```python
V = Function('V')(u,v)
V
```

```python
W = Function('W')(u,v)
W
```

```python
variables = Matrix( [u,v,x,y] )
variables
```

```python
differentials = Matrix( [du,dv,dx,dy] ) 
differentials
```

```python
lineElement = expand( -exp(-M)*du*dv + exp(-U) * ( cosh(W) * ( exp(V)*dx**2+ exp(-V)*dy**2 )- 2*sinh(W)*dx*dy) ) 
lineElement
```

```python
g = zeros(4)

for i in range(4):
    for j in range(4):
        if i == j:
            g[i,j] = lineElement.coeff( differentials[i],2 ) 
        else:
            g[i,j] = Rational(1,2)*lineElement.coeff( differentials[i]*differentials[j],1 )
            
g
```

```python
import sympy
```

```python
m = sympy.Array(g)
m
```

```python
from einsteinpy.symbolic import *
```

```python
syms = sympy.symbols("u v x y" , positive = True, real = True)
u,v,x,y = syms
```

```python
metric = MetricTensor(m,syms)
```

```python
ch = ChristoffelSymbols.from_metric(metric)
ch.tensor()
```

```python
Ric = RicciTensor.from_metric(metric)
Ric.tensor()
```

```python
R = RicciScalar.from_riccitensor(Ric)
R.simplify()
R.expr
```

```python
einst = EinsteinTensor.from_metric(metric)
einst.tensor()
```
