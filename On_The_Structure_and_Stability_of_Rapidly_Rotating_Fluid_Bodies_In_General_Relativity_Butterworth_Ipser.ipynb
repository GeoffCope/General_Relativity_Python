```python
t,r,theta,phi,dt,dr,dtheta,dphi = symbols('t r theta phi dt dr dtheta dphi')
```

```python
nu = Function('nu')(r,theta)
nu
```

```python
B = Function('B')(r,theta)
B
```

```python
omega = Function('omega')(r,theta)
omega
```

```python
zeta = Function('zeta')(r,theta)
zeta
```

```python
variables = Matrix( [t,r,theta,phi] )
variables
```

```python
differentials = Matrix( [dt,dr,dtheta,dphi] )
differentials 
```

```python
lineElement = expand(-exp(2*nu)*dt**2 + r**2 * sin(theta)**2 * B**2 * exp(-2*nu) * (dphi - omega * dt)**2 + exp(2*(zeta-nu))*(dr**2 + r**2 * dtheta**2) )
lineElement
```

```python
g = zeros(4)

for i in range(4):
    for j in range(4):
        if i == j:
            g[i,j] = lineElement.coeff( differentials[i],2 ) 
        else:
            g[i,j] = Rational(1,2)*lineElement.coeff( differentials[i]*differentials[j],1 )
            
g
```

```python
import sympy
```

```python
m = sympy.Array(g)
m
```

```python
Ric = RicciTensor.from_metric(metric)
Ric.tensor()
```

```python
R = RicciScalar.from_riccitensor(Ric)
R.simplify()
R.expr
```
